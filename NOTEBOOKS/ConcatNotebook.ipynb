```python
import pandas as pd
```

```python
df_2015 = pd.read_csv('./export_2015.csv')
df_2016 = pd.read_csv('./export_2016.csv')
df_2017 = pd.read_csv('./export_2017.csv')
df_2018 = pd.read_csv('./export_2018.csv')
df_2019 = pd.read_csv('./export_2019.csv')
df_2020 = pd.read_csv('./export_2020.csv')
```

```python
df_concat = pd.concat([df_2015, df_2016, df_2017, df_2018, df_2019, df_2020], axis=0)
```

```python
df_concat
```

```python
df_final = df_concat.drop(['Unnamed: 0', 'NumberofBuildings'], axis=1)
```
